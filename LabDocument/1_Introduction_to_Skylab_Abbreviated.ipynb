## EC2

In this section you will connect to your EC2 with VSCode Remote SSH. You should've been given an IP address, as well as a username and password for it. You'll use these to access your EC2 for the duration of this course. 

1. Open VSCode

2. Click on the "Remote Explorer" tab on the left column.

3. Select "SSH Targets" from the dropdown menu

<figure>

</figure>

4. Add a new connection by clicking on the plus sign (you'll see this once you hover your cursor over the "SSH TARGETS" line):


<figure>

</figure>

5. Login to your EC2 by entering your SSH command:

  ```
  ssh <USERNAME>@<IP_ADDRESS_OF_EC2>
  ```
  > **Note:** The ssh (Secure Shell) command has 3 parts: the ssh command, username, and host (IP/Domain name).

</figure>
